```python
import typing
import json
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
import numpy as np
import datetime
import time

%matplotlib inline
```

```python
def load_candles(path: str) -> typing.List[typing.Dict]:
    """Load candles from file

    Args:
        path (str): path to file

    Returns:
        typing.List[typing.Dict]: candles data
    """
    with open(path) as file:
        d = json.load(file)
        return d['candles']
```

```python
def plot_prices(candles: typing.List[typing.Dict]):
    candle_times = [datetime.datetime.strptime(c['time'], '%Y-%m-%dT%H:%M:%SZ') for c in candles]
    candle_open_values = [float(c['open']['units'] + '.' + str(c['open']['nano'])) for c in candles]
    candle_close_values = [float(c['close']['units'] + '.' + str(c['close']['nano'])) for c in candles]
    ax = plt.axes()
    ax.plot(candle_times, candle_open_values, color='red')
    ax.plot(candle_times, candle_close_values)
    ax.set_title('BR-6.32')
    ax.set_ylabel('Price')
    ax.set_xlabel('Time of Day')
    fmt = mdates.DateFormatter('%H:%M')
    ax.xaxis.set_major_formatter(fmt)
```

```python
candles = load_candles('../history/BR-6.23_2023-05-15.json')
plot_prices(candles)
```

```python
candles = load_candles('../history/BR-6.23_2023-05-16.json')
plot_prices(candles)
```
